```python
#Implement Insertion Sort 
#Lopp thorugh each index in range of the length of the list
#Check if the number at each index is greater than or less than the number to the index to the left
#If the number is greater than number to left, swap, and continue to decrease the index
#We make sure that the index is greater than 0.



def insert_sort(num_list):
    count=0 
    for i in range(1,len(num_list)):
        current=num_list[i]
        j=i
        while current<num_list[j-1] and j>=0:
            num_list[j]=num_list[j-1]
            count=count+1
            j=j-1
        num_list[j]=current
    print(num_list)
            
    
```
