Given: A positive integer nâ‰¤7.

Return: The total number of permutations of length n, followed by a list of all such permutations (in any order).

```python
from itertools import permutations
#Integer holds the number that is given
integer=7
#We use permutations from itertools to get a list of all the possible permutations of a number
#Add one to the number given due to zero indexing
permutation_list=list(permutations(range(1,integer + 1)))
#Count the number of permuations
print(len(permutation_list))
#Iterate throuh the permutations with map(), format output
for i in permutation_list:
        print(' '.join(map(str,i)))
```
